```python
#import libraries 
import pandas as pd
import numpy as np
```

#### Both comp_bundas_train and comp_bundas_train are datasets generated by the code written in R. 

These datasets have been modified in that the rows with missing data have either had their missing data filled from items with similar Item_IDs or have been removed.

```python
home_folder = '/Users/anthonymiyoro/Documents/code/DataTho/data'

comp_bundas_train = pd.read_csv(home_folder + 'comp_bundas_train.csv')
test_bundas = pd.read_csv(home_folder + 'comp_bundas_test.csv')
```

### Encode Categorical Variables on Test Data

```python
test_bundas.info()
```

```python
# One hot encode the categorical columns
FatContent_one_hot = pd.get_dummies(test_bundas['FatContent'], 
                                     prefix='FatContent')

Category_one_hot = pd.get_dummies(test_bundas['Category'], 
                                prefix='Category')

Store_Size_one_hot = pd.get_dummies(test_bundas['Store_Size'], 
                                   prefix='Store_Size')

# One hot encode the categorical columns
Store_Location_Type_one_hot = pd.get_dummies(test_bundas['Store_Location_Type'], 
                                     prefix='Store_Location_Type')

Store_Type_one_hot = pd.get_dummies(test_bundas['Store_Type'], 
                                prefix='Store_Type')

Store_ID_one_hot = pd.get_dummies(test_bundas['Store_ID'], 
                                   prefix='Store_ID')
```

```python
# Combine the one hot encoded columns with df_con_enc
modDfenc = pd.concat([test_bundas, 
                        FatContent_one_hot, 
                        Category_one_hot, 
                        Store_Size_one_hot,
                        Store_Location_Type_one_hot, 
                        Store_Type_one_hot, 
                        Store_ID_one_hot], axis=1)

# Drop the original categorical columns (because now they've been one hot encoded)
modDfenc = modDfenc.drop(['FatContent','Category','Store_Size','Store_Location_Type','Store_Type','Store_ID'], axis=1)
```

```python
modDfenc.head()
```

```python
modDfenc.info()
```

```python
modDfenc.to_csv(home_folder + 'test_modified.csv',index=False)
```
